```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def ndsolve(fs, init_values, t0, t1):
    n = 100
    output = [[t0, *init_values]]
    for i in range (1, (t1-t0)*n):
        t = t0+i/n
        output.append([t,*euler_next(fs,output[len(output)-1][1:],1/n)])
    return output

def euler_next(fs,x,deltat):
    out = [x[i] + deltat * fs[i](*x) for i in range (0, len(fs))]
    return out
    

fs = [lambda x,y: x - x**3 - y, lambda x,y: x ]
traj = ndsolve(fs, [2,1], 0, 30)
ttr, xtr, ytr = zip(*traj)
plt.plot(xtr, ytr)
plt.show()
```

```python
c1 = 0.15
c2 = 0.2
c3 = 7.1

fs = [lambda s1,s2,s3: -s2-s3, lambda s1,s2,s3: s1+c1*s2, lambda s1,s2,s3:c2+s3*(s1-c3)]
traj = ndsolve(fs, [0,0,0], 0, 100)

ts, xs, ys, zs = zip(*traj)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot(xs,ys,zs) # x􏰀値，y􏰀値，z􏰀値􏰀順
plt.show()
```
