```python
def nestlist(f,x0,n):
    output = [x0]
    for i in range (1, n):
        output.append(f(output[len(output)-1]))
    return output

def f_n(f, n, x0):
    temp = x0
    for i in range (0, n):
        temp = f(temp)
    return temp
```

```python
import random
import matplotlib.pyplot as plt

l = nestlist(lambda x: x + (-1)**random.randint(0, 1), 0, 1000)
plt.plot(l)
plt.show()
```
