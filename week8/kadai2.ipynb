```python
import matplotlib.pyplot as plt

remember_list = [0,0] #a, x0, n, x

def gen(a = 3.9, x0 = 0.1):
    global remember_list
    n = 0
    if a < 0 or a > 4:
        raise ValueError('0 ≤ a ≤ 4')
    elif x0 < 0 or x0 > 1.0:
        raise ValueError('0 ≤ x0 ≤ 1')
    
    if remember_list[0] == 0:
        x = x0
    else:
        n = remember_list[0]
        x = remember_list[1]
    
    for i in range (0, n):
        yield
        
    while True:
        xnext = a * x * (1 - x)
        yield xnext
        x = xnext
        n = n + 1
        remember_list = [n, x]
        
g = gen(a = 3.9, x0 = 0.1)
traj = [next(g) for i in range(100)]
plt.plot(traj)
plt.show()
```

```python
import matplotlib.pyplot as plt

remember_list = [0,0,0,0] #a, x0, n, x

def gen(a = 3.9, x0 = 0.1):
    global remember_list
    k = 0
    if a < 0 or a > 4:
        raise ValueError('0 ≤ a ≤ 4')
    elif x0 < 0 or x0 > 1.0:
        raise ValueError('0 ≤ x0 ≤ 1')
    
    if remember_list[0] != a or remember_list[1] != x0:
        remember_list = [a,x0,0,x0]
        k = remember_list[2]
        x = remember_list[3]
    else:
        k = remember_list[2]
        x = remember_list[3]
    
    for i in range (0, k):
        yield
        
    while True:
        xnext = a * x * (1 - x)
        yield xnext
        x = xnext
        k = k + 1
        remember_list = [a, x0, k, x]

xs = []
ys = []

n = 1000
m = 100

for i in range (0, n):
    for j in range (0, m):
        xs.append(i * 4 / n)

for i in range (0, n):
    g = gen(a = xs[i*m], x0 = 0.1)
    for j in range (0, m):
        ys.append(next(g))
        
plt.scatter(xs,ys,s=0.1)
```
