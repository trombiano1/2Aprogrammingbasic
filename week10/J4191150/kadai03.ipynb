```python
import math
import numbers

class Matrix:
    def __init__(self, a11, a12, a21, a22):
        self.a11 = a11
        self.a12 = a12
        self.a21 = a21
        self.a22 = a22
        
    def __repr__(self):
        s = f'[{self.a11} {self.a12}]\n'
        s += f'[{self.a21} {self.a22}]'
        return s
    
    def __add__(self, other):
        if(isinstance(self, numbers.Number)):
            ma1 = Matrix(self,self,self,self)
        else:
            ma1 = self

        if(isinstance(other, numbers.Number)):
            ma2 = Matrix(other,other,other,other)
        else:
            ma2 = other

        the_sum = Matrix(0,0,0,0)
        the_sum.a11 = ma1.a11 + ma2.a11
        the_sum.a12 = ma1.a12 + ma2.a12
        the_sum.a21 = ma1.a21 + ma2.a21
        the_sum.a22 = ma1.a22 + ma2.a22
        return the_sum
    
    def __sub__(self, other):
        if(isinstance(self, numbers.Number)):
            ma1 = Matrix(self,self,self,self)
        else:
            ma1 = self

        if(isinstance(other, numbers.Number)):
            ma2 = Matrix(other,other,other,other)
        else:
            ma2 = other
            
        return ma1.__add__(ma2.scalar(-1))
    
    def __mul__(self, other):
        if(isinstance(self, numbers.Number)):
            ma1 = Matrix(self,self,self,self)
        else:
            ma1 = self

        if(isinstance(other, numbers.Number)):
            ma2 = Matrix(other,other,other,other)
        else:
            ma2 = other
            
        the_product = Matrix(0,0,0,0)
        the_product.a11 = ma1.a11 * ma2.a11 + ma1.a12 * ma2.a21
        the_product.a21 = ma1.a21 * ma2.a11 + ma1.a22 * ma2.a21
        the_product.a12 = ma1.a11 * ma2.a12 + ma1.a12 * ma2.a22
        the_product.a22 = ma1.a21 * ma2.a12 + ma1.a22 * ma2.a22
        return the_product
    
    def __pow__(self,other):
        powerm = self
        for i in range (2, other+1):
            powerm = powerm * self
        return powerm
    
    def scalar(self, a):
        scalarproduct = Matrix(0,0,0,0)
        scalarproduct.a11 = a * self.a11
        scalarproduct.a12 = a * self.a12
        scalarproduct.a21 = a * self.a21
        scalarproduct.a22 = a * self.a22
        return scalarproduct
```

```python
print('m1')
m1 = Matrix(1,2,3,4)
print(m1)
print('m2')
m2 = Matrix(3,3,4,5)
print(m2)
print('add')
print(m1+10)
print('sub')
print(m1-10)
print('mul')
print(m1*10)
print('pow')
print(m1**3)
```
