### 計算

```python
import math
import numpy as np
from IPython.display import clear_output


def prepare(x_):
    ## PREPARE -----------------------
    #E, V, D_
    avg = np.mean(x_, axis=1)
    for j in range (0, len(x_)):
        for i in range (0, len(x_[0])):
            x_[j][i] -= avg[j]
    x = np.stack(tuple([a for a in x_])).T
    sigma = np.matrix(np.cov(x, rowvar=0, bias=1))
    eig_vals, E = np.linalg.eig(sigma)
    E_T = E.T
    D_ = np.diag(eig_vals**(-1/2))
    V = E * D_ * E_T
    #Z
    Z = np.dot(V,np.matrix(x.T))
    return Z,avg


## CALCULATE w ----------------------
def calculate_w(w,Z):
    threshold = 0.00001
    distance = 1
    while (threshold < distance):
        w_ = w.copy()
#       w = np.ravel(np.mean(np.dot(Z,np.diag(np.ravel(np.dot(w_,Z)))**3),axis = 1))-3*w_
#        w = np.ravel(np.mean(np.dot(Z,np.dot(w,Z).A1**3),axis = 1))-3*w
#        w[0] = np.mean(Z[:1].A1*np.dot(w,Z).A1**3) - 3 * w[0]
        w[0] = np.mean(Z[:1].A1*np.dot(w_,Z).A1**3) - 3*w_[0]
        w[1] = np.mean(Z[1:2].A1*np.dot(w_,Z).A1**3) -3*w_[1]
        w /= np.linalg.norm(w)
        distance = ((abs(w[0])-abs(w_[0]))**2+(abs(w[1])-abs(w_[1]))**2)**0.5
#        clear_output(wait=True)
#        print(threshold, "<", distance)
#    clear_output(wait=True)
    return w
    
## ICA ------------------------------
def ica(data):
    Z,avg = prepare(data)
    y = []
    for i in range (0, len(data)):
        w0 = [0.0 for i in range (0, len(data))]
        w0[i] = 1.0
        w = calculate_w(np.array(w0),Z)
        y_ = [0 for i in range (0, len(data[0]))]
        for i in range (0, len(data[0])):
           y_[i] = w[0]*np.asarray(Z)[0][i] + w[1]*np.asarray(Z)[1][i]
        y.append(y_)
    for j in range (0, len(y)):
        for i in range (0, len(y[0])):
            y[j][i] += avg[j]
    return y
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image1 = mpimg.imread('data/kadai4/image1.png')
image2 = mpimg.imread('data/kadai4/image2.png')

n = 512
m = 512

data_red = []
red = np.ravel([[image1[i][j][0] for j in range (0, 512)] for i in range (0, 512)])
data_red.append(red)
red = np.ravel([[image2[i][j][0] for j in range (0, 512)] for i in range (0, 512)])
data_red.append(red)
y_red = ica(data_red)

redmax = max(y_red[0])
redmin = min(y_red[0])
y_red = (y_red - (redmax + redmin)/2)/(redmax-redmin)+0.5

data_green = []
green = np.ravel([[image1[i][j][1] for j in range (0, 512)] for i in range (0, 512)])
data_green.append(green)
green = np.ravel([[image2[i][j][1] for j in range (0, 512)] for i in range (0, 512)])
data_green.append(green)
y_green = ica(data_green)

greenmax = max(y_green[0])
greenmin = min(y_green[0])
y_green = (y_green - (greenmax + greenmin)/2)/(greenmax-greenmin)+0.5

data_blue = []
blue = np.ravel([[image1[i][j][2] for j in range (0, 512)] for i in range (0, 512)])
data_blue.append(blue)
blue = np.ravel([[image2[i][j][2] for j in range (0, 512)] for i in range (0, 512)])
data_blue.append(blue)
y_blue = ica(data_blue)
bluemax = max(y_blue[0])
bluemin = min(y_blue[0])
y_blue = (y_blue - (bluemax + bluemin)/2)/(bluemax-bluemin)+0.5

image1 = [[np.array([y_red[0][i+num],y_green[0][i+num],y_blue[0][i+num]]) for num in range (0, m)] for i in range(0, n*m, m)]

plt.imsave("1.png",np.asarray(image1))
#plt.show()
```
