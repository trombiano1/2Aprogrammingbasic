```python
import numpy as np
import cmath
import matplotlib.pyplot as plt

xx = []
yy = []

for i in range (0, 100):
    arr = np.random.randn(200,200)/np.sqrt(200)
    m = np.asmatrix(arr)
    xx.append(np.linalg.eigvals(m).real)
    yy.append(np.linalg.eigvals(m).imag)

plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(xx, yy, s=0.1)
```
